```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def parse_data(data_file):
    new_list = []
    with open(data_file) as f:
        reader = csv.reader(f, delimiter = "\t")
        for row in reader:
            new_list.append(row)
    return new_list

gene_data = parse_data("C:\\Users\\browndo\\Documents\\brassica_flowering_expression.tsv")

def remove_blanks(input_list):
    new_list = [[w.replace("NA", "0") for w in l] for l in input_list]
    return new_list

def test_remove_blanks():
    a = [["2", "3", "NA"], ["4", "NA", "NA"]]
    if remove_blanks(a) == [["2", "3", "0"], ["4", "0", "0"]]:
        return True
    else:
        return False
    
test_remove_blanks()

gene_data = remove_blanks(gene_data)
#print(gene_data)
#gene_data_copy = [[w.replace("day_", "") for w in l] for l in gene_data]
```

```python
#given a list of data sets, plots them all on the same graph

exampleX = ["1", "2", "3"]
exampleY = [["2", "8", "16"], ["2", "3", "4"], ["2", "8", "16"], ["2", "3", "4"], ["2", "8", "16"], ["2", "3", "4"]]


def plot_data_together(listX, Y_master_list):
    lenY = int(len(Y_master_list[0]))
    for i in range(0, len(Y_master_list)):
        if (int(len(Y_master_list[i])) != lenY):
            print("Y value sets are not of same length")
            pass
    for i in range (0, len(Y_master_list)):
        plt.plot(listX, Y_master_list[i])
    plt.show()
```

```python
plot_data_together(exampleX, exampleY)
```

```python
#given a list of data sets, plots them on individual graphs

def plot_data_separate(listX, Y_master_list):
    fig = plt.figure(1, figsize = (20, 75))
    length = len(Y_master_list)
    plot_height = math.ceil(length / 2)
    for i in range(0, length):
        plt.subplot(plot_height, 2, (i + 1))
        plt.plot(listX, Y_master_list[i])
    plt.show()
```

```python
plot_data_separate(exampleX, exampleY)
```
